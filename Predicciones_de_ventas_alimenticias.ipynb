```python
#Función para asignación
def asignacion_pesos(dt_pesos_estimados_dic,dt_original_a_imputar):
  lista_pesos_asignados = []
  valor_promedio_pesos = dt_original_a_imputar['Item_Weight'].mean()
  for i in range(len(dt_original_a_imputar)):
    if pd.isna(dt_original_a_imputar['Item_Weight'][i]) == True:
      if pd.isna(dt_pesos_estimados_dic['Item_Weight'][dt_original_a_imputar['Item_Identifier'][i]]) == True:
        lista_pesos_asignados.append(valor_promedio_pesos)
      else:
          lista_pesos_asignados.append(dt_pesos_estimados_dic['Item_Weight'][dt_original_a_imputar['Item_Identifier'][i]])
    else:
      lista_pesos_asignados.append(dt_original_a_imputar['Item_Weight'][i])
  return lista_pesos_asignados
```

```python
data_predicciones['Item_Weight'] = asignacion_pesos(data_pesos_dict,data_predicciones)
```

```python
# Para poder asignar el tamaño se va a crear una función 
def funcion_imputar_tamanio_tienda(data_predicciones):
  lista_tamano_tienda = []
  for i in range(len(data_predicciones)):
      if data_predicciones['Outlet_Identifier'][i] == 'OUT010':
        lista_tamano_tienda.append('Small')
      elif (data_predicciones['Outlet_Identifier'][i] == 'OUT017') | (data_predicciones['Outlet_Identifier'][i] == 'OUT045'): 
        lista_tamano_tienda.append('Medium')
      else:
        lista_tamano_tienda.append(data_predicciones['Outlet_Size'][i])
  return lista_tamano_tienda      
```

**Identificar los valores actuales de hojas y profundidad**

```python
#La profundidad se obtiene así:
dec_tree.get_depth()
```

Ya sabemos que el modelo buscó los nodos más puros, llegando a una profundidad de 81 y un total de 6264 hojas; ahora para encontrar la mejor profundidad se puede crear una función para iterar entre 2 y la máx produndidad encontrada para hallar la profundidad donde está el mejor score de los datos de prueba

```python
# Lista de valores a probar para max_depth:
max_depth_range = list(range(2, 82))# will try every value between 2 and 82
# Lista para almacenar la puntuación para cada valor de max_depth:
r2 = []
rmse = []
depth_optimo = {}
depth_optimo_rmse = {}
for depth in max_depth_range:
    dec_tree = DecisionTreeRegressor(max_depth = depth, 
                             random_state = 42)
    dec_tree.fit(X_train_processed, y_train)
    score = dec_tree.score(X_test_processed, y_test)
    test_prediction = dec_tree.predict(X_test_processed)
    score_rmse = mean_squared_error(y_test, test_prediction,squared=False)
    r2.append(score)
    rmse.append(score_rmse)
    depth_optimo[score] = depth
    depth_optimo_rmse[score_rmse] = depth
```

```python
#Valor óptimo de la profundidad es para r2
val_depth_optimo_r2 = depth_optimo.get(max(r2))
val_depth_optimo_r2
```

```python
#Valor óptimo de la profundidad es para rmse
val_depth_optimo_rmse = depth_optimo_rmse.get(min(rmse))
val_depth_optimo_rmse
```

```python
dec_tree_2 = DecisionTreeRegressor(random_state=42, max_depth= val_depth_optimo_r2)
dec_tree_2.fit(X_train_processed,y_train)
train_prediction_2 = dec_tree_2.predict(X_train_processed)
test_prediction_2 = dec_tree_2.predict(X_test_processed)
r2_score_train_2 = r2_score(y_train, train_prediction_2)
r2_score_test_2 = r2_score(y_test, test_prediction_2)
print(f'R2 árbol optimizado en su profundidad, datos de entrenamiento: {r2_score_train_2}\nR2 árbol optimizado en su profundidad, datos de prueba: {r2_score_test_2}')
```

**Identificar los valores actuales de hojas y profundidad**

Ya sabemos que el modelo buscó los nodos más puros, llegando a una profundidad de 45 y un total de 6269 hojas; ahora para encontrar la mejor profundidad se puede crear una función para iterar entre 2 y la máx produndidad encontrada para hallar la profundidad donde está el mejor score de los datos de prueba

```python
# Lista de valores a probar para max_depth:
max_depth_range = list(range(2, 45))# will try every value between 2 and 45
# Lista para almacenar la puntuación para cada valor de max_depth:
r2 = []
rmse = []
depth_optimo = {}
depth_optimo_rmse = {}
for depth in max_depth_range:
    dec_tree = DecisionTreeRegressor(max_depth = depth, 
                             random_state = 42)
    dec_tree.fit(X_train_processed2, y_train2)
    score = dec_tree.score(X_test_processed2, y_test2)
    test_prediction = dec_tree.predict(X_test_processed2)
    score_rmse = mean_squared_error(y_test2, test_prediction2,squared=False)
    r2.append(score)
    rmse.append(score_rmse)
    depth_optimo[score] = depth
    depth_optimo_rmse[score_rmse] = depth
```

```python
#Valor óptimo de la profundidad es para r2
val_depth_optimo_r2_2 = depth_optimo.get(max(r2))
val_depth_optimo_r2_2
```

```python
#Valor óptimo de la profundidad es para rmse
val_depth_optimo_rmse_2 = depth_optimo_rmse.get(min(rmse))
val_depth_optimo_rmse_2
```

```python
dec_tree_3 = DecisionTreeRegressor(random_state=42, max_depth= val_depth_optimo_r2_2)
dec_tree_3.fit(X_train_processed2,y_train2)
train_prediction_3 = dec_tree_3.predict(X_train_processed2)
test_prediction_3 = dec_tree_3.predict(X_test_processed2)
r2_score_train_3 = r2_score(y_train2, train_prediction_3)
r2_score_test_3 = r2_score(y_test2, test_prediction_3)
print(f'R2 árbol optimizado en su profundidad, datos de entrenamiento: {r2_score_train_3}\nR2 árbol optimizado en su profundidad, datos de prueba: {r2_score_test_3}')
```
